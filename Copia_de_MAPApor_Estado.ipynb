```python
# ‚ñ∂Ô∏è Instalar dependencias
!pip install openpyxl geopy folium

# üìÅ Subir archivo
from google.colab import files
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import os
import zipfile

uploaded = files.upload()
archivo = next(iter(uploaded))  # Detectar autom√°ticamente el archivo subido

# üìö Leer archivo
df = pd.read_excel(archivo)
df = df.dropna(subset=['Latitud', 'Longitud'])

# üåç Geolocalizaci√≥n inversa
geolocator = Nominatim(user_agent="geo-tabletas-estados")
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)

coordenadas_unicas = df[['Latitud', 'Longitud']].drop_duplicates()
def obtener_estado(lat, lon):
    try:
        location = reverse((lat, lon), language='es')
        return location.raw['address'].get('state', 'Desconocido')
    except:
        return 'Desconocido'
coordenadas_unicas['Estado'] = coordenadas_unicas.apply(lambda row: obtener_estado(row['Latitud'], row['Longitud']), axis=1)
df = df.merge(coordenadas_unicas, on=['Latitud', 'Longitud'], how='left')

# üó∫Ô∏è Generar mapas por estado
output_folder = "mapas_html_por_estado"
os.makedirs(output_folder, exist_ok=True)

estados = df['Estado'].dropna().unique()
for estado in estados:
    df_estado = df[df['Estado'] == estado]
    if df_estado.empty: continue

    mapa = folium.Map(location=[df_estado.iloc[0]['Latitud'], df_estado.iloc[0]['Longitud']], zoom_start=7)
    cluster = MarkerCluster().add_to(mapa)

    for _, row in df_estado.iterrows():
        folium.Marker(
            location=[row['Latitud'], row['Longitud']],
            popup=f"<b>IMEI:</b> {row['IMEI/MEID']}<br><b>Serie:</b> {row['N√∫mero de serie']}",
            icon=folium.Icon(color='darkblue', icon='tablet', prefix='fa')
        ).add_to(cluster)

    nombre_archivo = f"Mapa_{estado.replace(' ', '_')}.html"
    mapa.save(os.path.join(output_folder, nombre_archivo))

# üì¶ Empaquetar en un ZIP
zip_path = "mapas_por_estado.zip"
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for root, dirs, files in os.walk(output_folder):
        for file in files:
            zipf.write(os.path.join(root, file), arcname=file)

# üì• Descargar
files.download(zip_path)
```

```python
# ‚ñ∂Ô∏è Instalar dependencias
!pip install openpyxl geopy folium

# üìÅ Subir archivo
from google.colab import files
import pandas as pd
import folium
from folium.plugins import MarkerCluster
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import os
import zipfile

# Store the uploaded files in a variable
uploaded_files = files.upload()
archivo = next(iter(uploaded_files))  # Detectar autom√°ticamente el archivo subido

# üìö Leer archivo
df = pd.read_excel(archivo)
df = df.dropna(subset=['Latitud', 'Longitud'])

# üåç Geolocalizaci√≥n inversa
geolocator = Nominatim(user_agent="geo-tabletas-estados")
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)

coordenadas_unicas = df[['Latitud', 'Longitud']].drop_duplicates()
def obtener_estado(lat, lon):
    try:
        location = reverse((lat, lon), language='es')
        return location.raw['address'].get('state', 'Desconocido')
    except:
        return 'Desconocido'
coordenadas_unicas['Estado'] = coordenadas_unicas.apply(lambda row: obtener_estado(row['Latitud'], row['Longitud']), axis=1)
df = df.merge(coordenadas_unicas, on=['Latitud', 'Longitud'], how='left')

# üó∫Ô∏è Generar mapas por estado
output_folder = "mapas_html_por_estado"
os.makedirs(output_folder, exist_ok=True)

estados = df['Estado'].dropna().unique()
for estado in estados:
    df_estado = df[df['Estado'] == estado]
    if df_estado.empty: continue

    mapa = folium.Map(location=[df_estado.iloc[0]['Latitud'], df_estado.iloc[0]['Longitud']], zoom_start=7)
    cluster = MarkerCluster().add_to(mapa)

    for _, row in df_estado.iterrows():
        folium.Marker(
            location=[row['Latitud'], row['Longitud']],
            popup=f"<b>IMEI:</b> {row['IMEI/MEID']}<br><b>Serie:</b> {row['N√∫mero de serie']}",
            icon=folium.Icon(color='darkblue', icon='tablet', prefix='fa')
        ).add_to(cluster)

    nombre_archivo = f"Mapa_{estado.replace(' ', '_')}.html"
    mapa.save(os.path.join(output_folder, nombre_archivo))

# üì¶ Empaquetar en un ZIP
zip_path = "mapas_por_estado.zip"
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for root, dirs, files_in_folder in os.walk(output_folder): # Changed 'files' to 'files_in_folder'
        for file in files_in_folder:
            zipf.write(os.path.join(root, file), arcname=file)

# üì• Descargar
files.download(zip_path) # Now using the original 'files' object from google.colab
```
